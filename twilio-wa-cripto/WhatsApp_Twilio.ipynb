## DataFrame

```python
def get_crypto(responseCrypto,i):
    nombre = responseCrypto['Data'][i]['CoinInfo']['FullName']
    precio = responseCrypto['Data'][i]['RAW']['USD']['PRICE']
    return nombre,precio
```

```python
datos = []
for i in tqdm(range(10), colour='green'):
    datos.append(get_crypto(responseCrypto,i))
```

```python
datos
```

```python
col = ['Nombre','Precio']
df = pd.DataFrame(datos, columns=col)
df
```

```python
decimals = pd.Series([2], index=['Precio'])
df.round(decimals)
```

```python
nuevo_df = df.round(decimals)
nuevo_df
```
